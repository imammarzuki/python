Contoh, deklarasikan variable a dengan suatu nilai, kemudian cek id dari objek a, nilai id yang tertulis pada contoh bisa berbeda dengan Anda.

```python
a=496
id(a)
```

Kemudian buat variable b dengan nilai lain, kemudian cek id nya, akan terlihat nilai id a dan b tidak sama.

```python
b=1729
id(b)
```

Kemudian kita buat statement bahwa b = a, lalu cek id variable b, maka id dari variable b akan sama dengan variable a.

```python
b=a
id(b)
```

```python
id(a)==id(b)
```

```python
a is b
```

Fungsi id() ini jarang digunakan pada aplikasi nyata, lebih sering digunakan pada tutorial seperti ini atau saat melakukan debugging,

Pada program, lebih umum menggunakan is operator untuk menguji equality of identity, menguji 2 objek mereferensi pada objek yang sama.

Jadi, identity dan value adalah 2 hal yang berbeda. 2 objek yang berbeda bisa memiliki value yang sama. Sedangkan identity menunjuk pada objek yang sama.

```python
a = [1, 2, 3]
b = [1, 2, 3]
a==b
```

```python
a is b
```
