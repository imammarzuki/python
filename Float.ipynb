Setiap angka dengan tanda decimal (.) atau dengan huruf e dikategorikan sebagai float oleh python.

Setiap operasi matematik antara tipe data float dengan integer akan diubah menjadi float.

```python
3.122
```

```python
3e8
```

```python
1.61e-35
```

```python
float(7)
```
