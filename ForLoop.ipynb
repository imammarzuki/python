Berikut contoh penggunaan for loop untuk menampilkan isi dari variable dengan type data dictionary (sering disebut sebagai associative array pada bahasa pemograman lain).

```python
listkota = ["Bandung", "Jakarta", "Denpasar"]
for kota in listkota:
  print(kota)
```

```python
ibukota = {"Jabar": "Bandung", "Jateng": "Semarang", "Jatim": "Surabaya"}
for provinsi in ibukota:
  print("Ibu kota Provinsi " + provinsi + " adalah " + ibukota[provinsi])
```
