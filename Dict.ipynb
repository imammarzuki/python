Untuk mendefinisikan dict gunakan fomat {k1: v1, k2: v2,…}, dengan k adalah key dan v adalah value

```python
d = {'Andi': '021-5757939', 'Ani': '023-12121030'}
```

Nilai dari dict dapat diakses dengan menggunakan key.

```python
d['Andi']
```

Karena dict adalah mutable, valuenya dapat diubah. Nilai dari dict[‘Andi’] diubah menjadi 022-57577939.

```python
d['Andi'] = '022-5757939'
```

```python
d
```

Untuk menambahkan isi dari dict, cukup dengan assign key dan value baru. Yang perlu diperhatikan dict baru akan ditambahkan, ordernya random.

```python
d['Anto'] = '062-12312312'
```

```python
d
```
