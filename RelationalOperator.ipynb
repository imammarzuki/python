Python Relational operator digunakan untuk membandingkan antar objek dan akan menghasilkan nilai bool. Pada bahasa programming lain, sering disebut boolean operator.

```python
g=100
```

```python
g==100
```

```python
g==13
```

```python
g!=13
```

```python
g>13
```

```python
g<13
```

```python
g<=100
```

```python
g>=100
```
