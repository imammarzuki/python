Python dict lanjutan akan membahas penggunaan dict lebih advance. Seperti dibahas pada lesson sebelumnya, dict adalah array key : value. Key harus uniqe, sementara value boleh duplikat(sama). syntax dari dict adalah : dict = {key1: value1, key2: value2, …}

Key object adalah immutable, jadi key boleh str, numbers, tuple tapi list tidak bisa menjadi key. Sementara value object adalah mutable.

***Perhatian** dict ordering adalah random, bahkan disetiap program dijalankan urutan isi dict bisa berbeda*

Constructor dict dapat digunakan untuk mengkonversi tuple dengan isi pair dari key dan value.

```python
a = [('Satu', 1), ('Dua', 2), ('Tiga', 3)]
```

```python
d = dict(a)
```

```python
d
```

```python
a = dict(Jan=1, Feb=2, Mar=3, Apr=4)
```

```python
a
```

# Iterasi pada dict

Gunakan for loop, iterasi pada key, gunakan d[key] untuk mengakses value. Dan perlu diperhatikan, pengurutan adalah sembarang. Kita juga melakukan iterasi untuk menampilkan value dari sebuah dict dengan fungsi .values()

```python
for key in c:
  print("{key} => {value}".format(key=key, value=c[key]))
```

```python
for value in c.values():
  print(value)
```

# Dict comprehension

```python
from pprint import pprint as pp
```

```python
country_capital = {'United Kingdom' : 'London', 'Brazil': 'Brazilia', 'Sweden': 'Stockholm'}
```

```python
capital_country = {capital: country for country, capital in country_capital.items()}
```

```python
pp(capital_country)
```
